```python
# =========================================================
# 🔷 BALANCED SUBSTRING: SINGLE CELL IMPLEMENTATION
# =========================================================

class BalancedSubstring:
    def __init__(self, x):
        self.x = x
        self.n = len(x)
        self.prefix = [0] * self.n

        # Build prefix sum (1 -> +1, 0 -> -1)
        val = 0
        for i in range(self.n):
            if x[i] == 1:
                val += 1
            else:
                val -= 1
            self.prefix[i] = val

    # -----------------------------------------------------
    # 🔹 Query: Check if substring [i, j] is balanced
    # Time: O(1)
    # -----------------------------------------------------
    def is_balanced(self, i, j):
        if i == 0:
            return self.prefix[j] == 0
        return (self.prefix[j] - self.prefix[i - 1]) == 0

    # -----------------------------------------------------
    # 🔹 Find Longest Balanced Substring
    # Time: O(N)
    # -----------------------------------------------------
    def longest_balanced(self):
        prefix = 0
        first_occurrence = {}
        max_length = 0

        for i in range(self.n):
            if self.x[i] == 1:
                prefix += 1
            else:
                prefix -= 1

            # Case 1: entire prefix balanced
            if prefix == 0:
                max_length = i + 1

            # Case 2: seen before
            if prefix in first_occurrence:
                length = i - first_occurrence[prefix]
                max_length = max(max_length, length)
            else:
                first_occurrence[prefix] = i

        return max_length


# =========================================================
# 🔷 USAGE EXAMPLE
# =========================================================

if __name__ == "__main__":
    x = [1, 0, 1, 1, 0, 0, 1]

    bs = BalancedSubstring(x)

    # 🔹 Query examples
    print("Is [1,4] balanced? ->", bs.is_balanced(1, 4))  # True
    print("Is [0,3] balanced? ->", bs.is_balanced(0, 3))  # False

    # 🔹 Longest balanced substring
    print("Longest balanced length ->", bs.longest_balanced())  # 6
```
